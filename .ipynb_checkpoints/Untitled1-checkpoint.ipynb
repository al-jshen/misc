```python
%autoreload

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from joblib import Parallel, delayed
from tqdm.notebook import tqdm
import warnings
warnings.filterwarnings("ignore")
import matplotlib
import cmasher as cmr

plt.rcParams['axes.labelsize'] = 25
plt.rcParams['axes.titlesize'] = 25
plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 25
plt.rcParams['figure.figsize'] = (11, 7)
plt.rc('text', usetex=True)
```

```python
rate = .013
invest = 0.04
years = 25

cibc_price = 568900
cibc_down = 0.35 * cibc_price
cibc_cash = 0

td_price = 583900
td_down = 0.2 * td_price
td_cash = cibc_down - td_down
```

```python
def mortgage(amount_left, years, rate):
    n = years * 12
    r = rate / 12
    with np.errstate(divide='ignore',invalid='ignore'):
        return amount_left * ( r * (1 + r)**n ) / ( (1 + r)**n - 1)
```

```python
def investment(initial, years, rate):
    n = years * 12
    r = rate / 12
    return initial * (1 + r)**n
```

```python
def run_mortgage_with_investment(prop_value, down, initial_investment, years, interest_rate, investment_rate, cutoff_years=None):
    al = prop_value - down
    y = years
    r = interest_rate
    ii = initial_investment

    payments = 0

    for i in range(years * 12):
        al = investment(al, 1/12, interest_rate)
        m = mortgage(al, y, r)
        al -= m
        y -= 1/12
        payments += m
        ii = investment(ii, 1/12, investment_rate)

        if cutoff_years:
            if i == cutoff_years * 12:
                break

    total_paid = payments + down
    
    return al, payments, total_paid, ii, total_paid - ii + al
```

```python
run_mortgage_with_investment(cibc_price, cibc_down, cibc_cash, years, rate, invest, 10)
```

```python
run_mortgage_with_investment(td_price, td_down, td_cash, years, rate, invest, 10)
```

```python
def cibc_over_td(cibc_price, cibc_down, cibc_cash, td_price, td_down, td_cash, years, rate, invest, cutoff_years=None):
    cibc = run_mortgage_with_investment(cibc_price, cibc_down, cibc_cash, years, rate, invest, cutoff_years=cutoff_years)
    td = run_mortgage_with_investment(td_price, td_down, td_cash, years, rate, invest, cutoff_years=cutoff_years)
    return cibc[-1] - td[-1]
```

```python
all_rates = np.linspace(0., 0.2)
all_invest = np.linspace(0., 0.2)
rr, ii = np.meshgrid(all_rates, all_invest)

diffs = []
for i in tqdm(range(rr.shape[0])):
    diffs.append(
        np.array(
            Parallel(n_jobs=-1)(delayed(cibc_over_td)(cibc_price, cibc_down, cibc_cash, td_price, td_down, td_cash, years, rj, ij, cutoff_years=10) for rj, ij in zip(rr[i], ii[i]))
        )
    )
diffs = np.array(diffs)
```

```python
plt.pcolormesh(rr, ii, diffs / 1e4, cmap=cmr.iceburn, norm=matplotlib.colors.TwoSlopeNorm(0), shading='gouraud')
plt.colorbar(label='TD saved compared to CIBC (\$10k)')
plt.contour(rr, ii, diffs / 1e4, levels=[0], colors='w')
# plt.colorbar()
plt.xlabel('Mortgage Rate')
plt.ylabel('Investment Return')
```
