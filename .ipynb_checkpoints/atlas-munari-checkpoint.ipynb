```python
import faiss
import matplotlib.pyplot as plt
import numpy as np
import opt_einsum as oe
import pandas as pd
from astropy.io import fits
from prfr import split_arrays
```

```python
f = fits.open("atlas9-munari.hires.grid.fits")
```

```python
wl = f[0].data[-1]
spec = f[0].data[:-1]
params = pd.DataFrame(f[1].data).to_numpy()[:, :4]
```

```python
randomizer = np.random.permutation(spec.shape[0])
train, test = split_arrays(spec[randomizer], params[randomizer], test_size=0.1)
trainspec, trainparams = train
testspec, testparams = test
# mask = params[:, 1] == 2
# testspec = spec[mask]
# testparams = params[mask]
# trainspec = spec[~mask]
# trainparams = spec[~mask]
```

```python
m = trainspec.mean(axis=0)
s = trainspec.std(axis=0)
```

```python
M = 128
ef_search = 128
ef_construction = 128
neighbours = 5
index = faiss.IndexHNSWFlat(wl.size, M)
index.hnsw.efSearch = ef_search
index.hnsw.efConstruction = ef_construction
index.add((trainspec - m) / s)
```

```python
D, I = index.search((testspec.reshape(-1, wl.size) - m) / s, 5 + 1)
w = 1 / (D + np.sqrt(1e-7))
w /= w.sum(axis=1)[:, None]
pred = oe.contract("ijk, ij -> ik", trainparams[I], w)
```

```python
fig, ax = plt.subplots(1, 4, figsize=(20, 4))
for i in range(4):
    ax[i].scatter(testparams[:, i], pred[:, i], s=5, alpha=0.8)
    ax[i].plot(testparams[:, i], testparams[:, i], c="r", zorder=5)
    ax[i].annotate(
        np.std(testparams[:, i] - pred[:, i]), xy=(0.1, 0.8), xycoords="axes fraction"
    )
```
