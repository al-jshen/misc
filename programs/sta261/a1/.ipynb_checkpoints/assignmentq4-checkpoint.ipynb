## part f

The size that $n$ needs to be in order for $\overline{X}$ to converge to normal depends on the initial distribution, so it is hard to make a comment on it that applies to all cases. An extremely rough estimate might be $n=30$. The more skewed the original distribution is, the larger $n$ needs to be. For example, comparing the graphs in part b and part c, even though $n=5$ in both cases, the graph in part b looks closer to a normal distribution than the one in part c because the one in part c is very skewed. This is because the original distribution of part c ($\chi_{df=2}^2$) is skewed too.
