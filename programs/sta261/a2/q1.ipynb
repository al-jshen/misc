## part b

```python
#(i)
def gentest():
    samp = np.random.normal(5, 3, 10)
    l_theta0 = np.prod(1 / np.sqrt(2 * np.pi * 9) * np.exp( - np.square(samp - 5) / (2 * 9)))
    l_theta1 = np.prod(1 / np.sqrt(2 * np.pi * 9) * np.exp( - np.square(samp - samp.mean()) / (2 * 9)))
    return -2 * np.log(l_theta0 / l_theta1)
```

```python
#(ii)
lrt_vec = np.zeros(100000)
for i in np.arange(100000):
    lrt_vec[i] = gentest()
lrt_vec
```

```python
#(iii)
plt.hist(lrt_vec, bins=100, density=True, label='histogram from normal data');
plt.legend();
```

```python
#(iv)
plt.hist(lrt_vec, bins=100, density=True, label='histogram from normal data');
sns.distplot(np.random.chisquare(df=1, size=100000), hist=False, label='chi squared')
```

## part c

```python
#(i)
def gentest():
    samp = np.random.poisson(5, 10)
    l_theta0 = np.prod(np.power(5, samp) * np.exp(-5) / factorial(samp))
    l_theta1 = np.prod(np.power(samp.mean(), samp) * np.exp(-samp.mean()) / factorial(samp))
    return -2 * np.log(l_theta0 / l_theta1)
```

```python
#(ii)
lrt_vec = np.zeros(100000)
for i in np.arange(100000):
    lrt_vec[i] = gentest()
lrt_vec
```

```python
#(iii)
plt.hist(lrt_vec, bins=100, density=True, label='histogram from poisson data');
plt.legend();
```

```python
#(iv)
plt.hist(lrt_vec, bins=100, density=True, label='histogram from poisson data');
sns.distplot(np.random.chisquare(df=1, size=100000), hist=False, label='chi squared')
```

I expect that as the sample size increases, the histogram becomes closer to the chi squared density. This is because the Central Limit Theorem says that these distributions should converge to normal, which means that the squares should converge to chi-squared.
