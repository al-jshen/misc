```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
def run_trials(trials, total_money, A_MONEY, a_prob):
    a_money = A_MONEY
    b_money = B_MONEY = total_money - a_money
    logs = []
    for i in range(trials):
        counter = 0
        while a_money != total_money and b_money != total_money:
            round_result = np.random.choice(2, p=(a_prob, 1-a_prob))
            if round_result == 0:
                a_money += 1
                b_money -= 1
            else:
                a_money -= 1
                b_money += 1
            counter += 1
        winner = '0' if a_money == total_money else '1'
        logs.append((winner, counter))
        a_money = A_MONEY
        b_money = B_MONEY
    return np.array(logs, dtype='int')
```

```python
# given a fair game, rates of winning should be 50/50

TOTAL_MONEY = 10
A_PROB = 0.5
A_MONEY = 5
BET = 1
logs = run_trials(10000, TOTAL_MONEY, A_MONEY, A_PROB)
sns.countplot(logs[:,0])
plt.show()
sns.boxplot(logs[:,1])
```

```python
# given different starting money, A should win a_money/total_money of the time for a fair game
TOTAL_MONEY = 10
A_PROB = 0.5
A_MONEY = 7
BET = 1
logs = run_trials(10000, TOTAL_MONEY, A_MONEY, A_PROB)
sns.countplot(logs[:,0])
plt.show()
sns.boxplot(logs[:,1])
```

```python
# given an unfair game, A should win ((1-((1-A_PROB)/A_PROB)**(A_MONEY)) / (1-((1-A_PROB)/A_PROB)**(TOTAL_MONEY)))

TOTAL_MONEY = 10
A_PROB = 0.7
A_MONEY = 1
BET = 1
logs = run_trials(10000, TOTAL_MONEY, A_MONEY, A_PROB)
sns.countplot(logs[:,0])
plt.show()
sns.boxplot(logs[:,1])
print((1-((1-A_PROB)/A_PROB)**(A_MONEY)) / (1-((1-A_PROB)/A_PROB)**(TOTAL_MONEY)))
```
