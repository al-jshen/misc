```python
import re

import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/Users/js5013/Desktop/response.csv")
```

```python
df = df.drop("Timestamp", axis=1)
```

```python
extract_name = lambda s: re.sub(r".*\[(.*)\]", r"\g<1>", s)
```

```python
df.columns = [extract_name(c) for c in df.columns]
```

```python
df_fmt = df.apply(df.value_counts).fillna(0).T
df_fmt.to_clipboard()
df_fmt
```

```python
plt.imshow(df_fmt / df_fmt.sum(axis=0))
```
